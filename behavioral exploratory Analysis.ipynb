```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Excluding the 'default_payment_next_month' column
credit_data_vif = credit_data2.drop(['default payment next month'], axis=1)

# constant (intercept) term to the model
credit_data_vif = sm.add_constant(credit_data_vif)

# VIF for each feature
vif_data = pd.DataFrame()
vif_data["Feature"] = credit_data_vif.columns
vif_data["VIF"] = [variance_inflation_factor(credit_data_vif.values, i) for i in range(credit_data_vif.shape[1])]

print(vif_data)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your dataset is already loaded into credit_data2

# Create age groups (bins and labels)
bins = [20, 29, 39, 49, 59, float('inf')]  # Create age bins
labels = ['20-29', '30-39', '40-49', '50-59', 'Above 60']  # Labels for the age groups
credit_data2['AGE_GROUP'] = pd.cut(credit_data2['AGE'], bins=bins, labels=labels, right=True)

# Create a function to generate stacked bar charts with annotations
def plot_default_rates(ax, group_column, labels, title):
    # Calculate counts for each class (0 and 1) in each group
    group_counts = credit_data2.groupby([group_column, 'default payment next month']).size().unstack()

    # Plot the results (absolute counts)
    group_counts.plot(kind='bar', stacked=True, ax=ax, color=['lightblue', 'coral'])
    
    # Set title and labels
    ax.set_title(title)
    ax.set_xlabel(labels[0])
    ax.set_ylabel('Count')
    ax.set_xticks(range(len(group_counts.index)))
    ax.set_xticklabels(group_counts.index, rotation=0)
    ax.legend(['Non-Default (0)', 'Default (1)'], title='Class')

    # Annotate the bars with the correct numbers inside each section
    for p in ax.patches:
        height = p.get_height()  # Get height of each bar segment
        if height > 0:
            ax.annotate(f'{int(height)}', 
                        (p.get_x() + p.get_width() / 2, p.get_y() + height / 2), 
                        ha='center', va='center', xytext=(0, 5), textcoords='offset points')

    # Add total counts at the top of the bars
    totals = group_counts.sum(axis=1)
    for i, total in enumerate(totals):
        ax.annotate(f'{int(total)}', 
                    (i, total), 
                    ha='center', va='bottom', xytext=(0, 5), textcoords='offset points')

# Create a figure with subplots
fig, axs = plt.subplots(2, 2, figsize=(15, 12))

# Define the columns and titles for each subplot
plots_info = [
    ('AGE_GROUP', ['Age Group'], 'Default Rates by Age Groups'),  # Updated to 'AGE_GROUP'
    ('MARRIAGE', ['Marital Status'], 'Default Rates by Marital Status'),
    ('SEX', ['Gender'], 'Default Rates by Gender'),
    ('EDUCATION', ['Education Level'], 'Default Rates by Education Level')
]

# Generate each plot in its respective subplot
for ax, (column, labels, title) in zip(axs.flatten(), plots_info):
    # Create a readable column for Marital Status and Education Level
    if column == 'MARRIAGE':
        credit_data2['MARITAL_STATUS'] = credit_data2['MARRIAGE'].map({1: 'Married', 2: 'Single', 3: 'Others'})
        plot_default_rates(ax, 'MARITAL_STATUS', labels, title)
    elif column == 'EDUCATION':
        credit_data2['EDUCATION_LEVEL'] = credit_data2['EDUCATION'].map({1: 'Graduate School', 2: 'University', 3: 'High School', 4: 'Others'})
        plot_default_rates(ax, 'EDUCATION_LEVEL', labels, title)
    else:
        plot_default_rates(ax, column, labels, title)

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your dataset is already loaded into credit_data2

# Create a function to generate stacked bar charts with annotations
def plot_default_rates(ax, group_column, labels, title):
    # Calculate counts for each class (0 and 1) in each group
    group_counts = credit_data2.groupby([group_column, 'default payment next month']).size().unstack()

    # Plot the results (absolute counts)
    group_counts.plot(kind='bar', stacked=True, ax=ax, color=['lightblue', 'coral'])
    
    # Set title and labels with increased fontsize
    ax.set_title(title, fontsize=14)  # Increase title font size
    ax.set_xlabel(labels[0], fontsize=14)  # Increase x-axis label font size
    ax.set_ylabel('Count', fontsize=14)  # Increase y-axis label font size
    ax.set_xticks(range(len(group_counts.index)))
    ax.set_xticklabels(group_counts.index, rotation=0, fontsize=12)  # Increase tick label font size
    ax.legend(['Non-Default (0)', 'Default (1)'], title='Class', prop={'size': 14})  # Increase legend font size

    # Annotate the bars with the correct numbers inside each section
    for p in ax.patches:
        height = p.get_height()  # Get height of each bar segment
        if height > 0:
            # If the height is too small, adjust annotation position to avoid overlap
            if height < 5:
                ax.annotate(f'{int(height)}', 
                            (p.get_x() + p.get_width() / 2, p.get_y() + height + 1),  # Move above small bars
                            ha='center', va='center', xytext=(0, 5), textcoords='offset points', fontsize=10)
            else:
                ax.annotate(f'{int(height)}', 
                            (p.get_x() + p.get_width() / 2, p.get_y() + height / 2), 
                            ha='center', va='center', xytext=(0, 5), textcoords='offset points', fontsize=10)

    # Add total counts at the top of the bars
    totals = group_counts.sum(axis=1)
    for i, total in enumerate(totals):
        ax.annotate(f'{int(total)}', 
                    (i, total), 
                    ha='center', va='bottom', xytext=(0, 5), textcoords='offset points', fontsize=10)

# Create a figure with subplots
fig, axs = plt.subplots(2, 2, figsize=(15, 12))

# Define the columns and titles for each subplot
plots_info = [
    ('AGE_GROUP', ['Age Group'], 'Default by Age Groups'),
    ('MARRIAGE', ['Marital Status'], 'Default by Marital Status'),
    ('SEX', ['Gender'], 'Default by Gender'),
    ('EDUCATION', ['Education Level'], 'Default by Education Level')
]

# Generate each plot in its respective subplot
for ax, (column, labels, title) in zip(axs.flatten(), plots_info):
    # Create readable columns for Marital Status, Education Level, and Gender
    if column == 'MARRIAGE':
        credit_data2['MARITAL_STATUS'] = credit_data2['MARRIAGE'].map({1: 'Married', 2: 'Single', 3: 'Others'})
        plot_default_rates(ax, 'MARITAL_STATUS', labels, title)
    elif column == 'EDUCATION':
        credit_data2['EDUCATION_LEVEL'] = credit_data2['EDUCATION'].map({1: 'Graduate School', 2: 'University', 3: 'High School', 4: 'Others'})
        plot_default_rates(ax, 'EDUCATION_LEVEL', labels, title)
    elif column == 'SEX':
        credit_data2['GENDER'] = credit_data2['SEX'].map({1: 'Female', 2: 'Male'})
        plot_default_rates(ax, 'GENDER', labels, title)
    else:
        plot_default_rates(ax, column, labels, title)

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```
